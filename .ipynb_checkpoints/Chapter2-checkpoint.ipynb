### The While Loop

Used to execute a set of statements as long as a condition is true

```python
# Example
i = 0
while i < 10:
    print(i)
    
    # Make sure to increment the initial value to prevent your code from running forever
    i += 1
```

### The 'break' and 'continue' statements 

break statement is used to stop the loop even if the condition is true.

the continue statement stops the current iteration and proceeds to the next.

A while loop can also have an else statement.

```python
# Example

i = 0
while i < 10:
    if i == 5:
        print("Don't show this one...")
        i += 1
        continue
    else:
        print(i)
        i += 1
```

- It takes a number of arguments but have one expression.

#### OBJECT METHODS (Methods are functions that belong to the object)

```python
class Cars:
    def __init__(self, tyres, doors, seats, boot, mirrors):
        self.tyres = tyres
        self.doors = doors
        self.seats = seats
        self.boot = boot
        self.mirrors = mirrors
    def details(self):
        print("The car has", self.tyres, "tyres,", self.doors, "doors,", self.seats, "seats", self.boot, "boot and", self.mirrors, "mirrors.")
```

```python
car = Cars(16, 2, 2, 1, 2)
```

```python
car.details()
```

```python
del car.boot
car.boot
```

```python
del car
car.doors
```

### INHERITANCE

```python
# Allows us to define a class that inherits all the methods and properties from another class.
# Parent class is the class being inherited from.
# Child class is the class thAT INHERITS FROM ANOTHER CLASS.
```

```python
class NewCars(Cars):
    def __init__(self, gears):
        self.gears = gears
        super().__init__(tyres, doors, seats, boot, mirrors)
```

```python
CarModel = NewCars(8,2,2,1,2,1)
CarModel.tyres
```

### Write the following sentence in a file called Program3.txt.
- This class has been a bit difficult but educative.
- First check if the file exists before writing to it, else create the file and write to it.

```python
if os.path.exists("Program3.txt"):
    file = open("Program3.txt", "w")
    file = file.write("This class has been a bit difficult but educative.")
else:
    file = open("Program3.txt", "x")
    file = open("Program3.txt", "w")
    file = file.write("This class has not been difficult but educative.")
    file = file.close()
```
